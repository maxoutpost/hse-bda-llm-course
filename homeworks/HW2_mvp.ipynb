# ДЗ2. Проект (MVP) или воспроизведение статьи

## Введение

Проект (или воспроизведение статьи) - это альтернативная траектория вместо соревнования: вы не гонитесь за местом в лидерборде, а собираете свой MVP-прототип LLM-системы (RLHF / RAG / агенты - что вам ближе) или аккуратно воспроизводите научную работу.

Такой формат особенно подходит тем, у кого ВКР/интересы связаны с NLP/LLM и хочется "потрогать" идею, получить валидацию и фактически сделать мини-предзащиту; тем, кому тесно в жестких рамках соревнования и важна свобода в выборе темы, данных, способов SFT/alignment, метрик валидации и простого UI; а также тем, кто хочет целенаправленно потренировать alignment и устойчивость к типовым хакам в более прикладном сценарии.


Вы выбираете один из двух треков:

- **Трек 1. MVP-проект.** Придумать задачу, обосновать данные, собрать работающий прототип LLM/NLP-системы (без сервера/фронта), оценить качество и устойчивость.
- **Трек 2. Воспроизведение статьи.**
Выбрать статью по LLM/NLP и *самостоятельно* воспроизвести ключевой метод/эксперименты в масштабе, доступном на Colab/Kaggle/API.        

**Согласование:**

Трек 2 - **обязательно** согласовать статью с [преподавателем](https://github.com/AlbinaBurlova) до начала реализации.

Трек 1 - согласование не требуется, но приветствуется короткое сообщение с темой и данными: чтобы вовремя остановить слишком большой проект.


## Трек 1 - MVP-проект


### Идея и ограничения

Вы предлагаете собственную задачу, в которой ключевую роль играет LLM или методы NLP, и делаете *минимально жизнеспособный прототип* (MVP) - ровно до точки, где видно:

1. что задача корректно поставлена,
2. есть данные и измеримая проверка,
3. пайплайн работает,
4. понятно, где ломается и почему.

Серверная часть, фронтенд, продакшн-обвязка, деплой, мониторинг **не требуются** - фокус на содержательном ядре: данные -> модель/система -> оценка -> диагностика.

### Постановка задачи и критерий успеха

Сначала сформулируйте задачу так, чтобы её можно было понять без контекста: какую проблему вы решаете, для кого это важно и почему именно LLM/NLP здесь уместен (а не “обычная” модель или правила).

Дальше зафиксируйте контракт системы: что является входом и что считается корректным выходом - например, "на вход подаём текст обращения, на выходе получаем JSON со слотами" или "на вход вопрос, на выход ответ с опорой на источники".

И наконец, выберите один главный критерий успеха: одну основную метрику, по которой вы будете судить, стало ли лучше, и коротко объясните, почему она соответствует вашей задаче (например, accuracy/F1 для классификации, exact match или качество извлечения полей для extraction, faithfulness + relevancy для RAG-ответов и т. п.).


### Данные

Дальше опишите данные так, чтобы было ясно, что вы контролируете качество входа в систему. Укажите источник, доступность и ограничения: лицензия, язык, возможные смещения, шум, чувствительность данных (если актуально).

Затем коротко расскажите, как вы готовили данные: какие шаги очистки делали, как приводили все к нужному формату, как делили на train/dev/test (или на базу знаний + тест-кейсы для RAG). В репозитории сохраните артефакты вашего разведовочного анализа данных.

Обязательно приведите 5-10 живых примеров из данных с комментариями - это лучший способ показать, что вы понимаете, что именно пытаетесь решить и где находятся подводные камни.

> Реалистичный масштаб данных для MVP: маленький корпус/датасет, который вы действительно сможете понять и руками проверить.

### MVP-реализация

В этой части вам нужно собрать работающий прототип. В работе обязательно должны быть две версии решения: **baseline** - самый простой, но разумный вариант, который уже решает задачу; и **осмысленное улучшение**, которое вы добавляете поверх baseline, объясняя, зачем оно нужно и какую проблему оно закрывает.

> Важно: лучше одно-два улучшения по делу, чем десять поверхностных абляций.

Дальше вы выбираете основной тип решения (можно комбинировать, если это действительно нужно задаче). Например, вы можете построить **RAG-систему**: подготовить корпус, разбить его на фрагменты, настроить embeddings и поиск, затем спроектировать промпт и добиться ответов с привязкой к источникам, чтобы было видно, откуда взялась информация.

Или вы можете пойти в сторону **адаптации модели** под вашу задачу: подготовить данные в нужном формате, сделать SFT/LoRA/prompt tuning, затем оценить качество и разобрать ошибки - где модель справляется, а где начинает ломаться.

Ещё одна допустимая траектория - **alignment на предпочтениях**: вы формируете пары chosen/rejected (минимальный масштаб - порядка сотни осмысленных пар), выбираете простой и реалистичный для MVP метод обучения по предпочтениям (например, DPO/ORPO, или упрощенный RLHF-вариант), после чего сравниваете поведение модели до/после на отложенных примерах и анализируете, какие типовые отказы появились или исчезли.

Если ваша задача больше про *систему*, чем про обучение, можно сделать **промпт-пайплайн или tool-use**: продумать цепочку шагов, подключить один-два инструмента (например, поиск по базе/простую валидацию/структурирование), задать строгий формат ответа и показать на примерах, где именно пайплайн устойчив, а где ломается и почему.

При этом все эти подходы можно **разумно совмещать**: например, сделать RAG как основу, а поверх добавить промпт-пайплайн (с этапом структурирования и валидации ответа); или сначала собрать baseline на prompting, а затем улучшить его LoRA-адаптацией; или обучить модель под стиль/тон (SFT/DPO), но все равно использовать retrieval, чтобы ответы были опирающимися на источники.

Главное - показать понятную логику: что является базовым решением, что именно вы добавили сверху и какую измеримую проблему это улучшение решает.


### Оценка качества и валидация устойчивости

LLM-системы особенно легко съезжают по качеству из-за того, что для них размыта граница **инструкция vs данные**: модель может воспринимать текст из запроса или из контекста как команду, а не как информацию. Поэтому prompt injection - это не частный случай, а типовой риск, который нужно учитывать в оценке.

В этом ДЗ важно показать **разумную валидацию**: вы заранее определяете, как проверяете качество, где ожидаете отказы, и демонстрируете это на тестах. То есть вы валидируете не только "средний сценарий", но и устойчивость к краевым случаям и простым попыткам обойти правила.

**Базовая оценка качества: тест-набор и контроль качества**

Соберите **тестовый набор 20-25 кейсов** (минимум), который отражает ваш реальный сценарий, и обязательно добавьте туда несколько сложных/краевых примеров. Для этих кейсов зафиксируйте ожидаемое поведение: что считается успехом и что провалом - и покажите хотя бы один автоматический или полуавтоматический контроль качества: например, проверку формата, проверку опоры на контекст для RAG, сравнение с эталоном/известными фактами или LLM-as-judge.

Если у вас RAG-система, особенно приветствуется раздельная оценка того, *что нашлось* (retrieval) и того, *что сгенерировалось* (generation): например, насколько контекст релевантен запросу и насколько ответ действительно опирается на найденные фрагменты, а не галлюцинирует.

**Валидация устойчивости: негативные кейсы**

Отдельно прогоните систему на **3-5 атакующих/негативных кейсах** и покажите результаты. Это про честную проверку того, что ваша система ведёт себя предсказуемо в ситуациях, где она обычно ломается. В набор можно включить: prompt injection в пользовательском запросе (попытка переопределить роль/политику или заставить нарушить правила), инъекцию внутри документов (для RAG), конфликт источников, недостаток контекста (где корректная реакция - уточнить/отказаться/обозначить неопределённость), а также форматный хак, когда модель пытается уйти из требуемого формата.

Минимальный порог здесь простой: 3-5 примеров с наблюдаемым поведением и выводами. Главное - показать, что вы это проверяли и понимаете, какие риски остаются.

**Если вы используете LLM-as-judge**

LLM-as-a-judge разрешён, но его нельзя использовать как единственную оценку. Обязательно добавьте хотя бы один контроль, который не зависит от judge-модели (детерминированная проверка формата/ссылок/источников/схемы), и учитывайте, что judge тоже может быть уязвим к промпт-инъекциям и манипуляциям формулировками.

**Если RAG**

Если у вас retrieval-контекст, зафиксируйте (в промпте и в логике пайплайна) принцип: retrieved контент - это данные, а не инструкции. В оценке это должно проявляться так, что на инъекционных документах система не выполняет команды из контекста, а продолжает отвечать по задаче, ссылаясь на источники и сохраняя требуемый формат. Это базовая практическая рекомендация против prompt injection и хороший тест на корректность дизайна RAG-пайплайна.

### Формат сдачи

Чтобы ваш проект выглядел как настоящий законченный MVP, просим сдавать его аккуратно упакованным в репозитории - так, чтобы по нему было легко понять идею, быстро запустить код и увидеть результат. Плюс: такой репозиторий можно потом спокойно показывать как портфолио-проект в резюме - с понятной структурой, отчетом, демо и воспроизводимыми экспериментами.

0. **Оформление в репозитории.** Работа должна быть упакована как мини-проект: отдельная папка, понятные названия файлов, короткий `README.md` с описанием задачи, структуры проекта и инструкцией "как запустить", фиксированные зависимости (`requirements.txt`), а также (если применимо) папка `data/` или ссылка на данные.

1. **Ноутбук/ноутбуки** `.ipynb` с EDA, кодом, результатами и пояснениями.

2. **Краткий отчет** (в ноутбуке или `report.md`). Задача, данные, baseline, улучшение, метрики, ошибки, верификация/атаки, план развития.

3. **Данные/инструкции**. Либо приложить выборку, либо ссылка + скрипт/инструкция воспроизведения.

4. *(желательно)* `slides.pdf` для защиты.


### Защита

За 10 минут вам нужно полно изложить:

1. задачу и её мотивацию;
2. охарактерировать данные (EDA) и подход;
3. демонстрация  MVP;
4. верификация + выводы;
5. план развития проекта.

После презентации последует серия вопросов по содержанию выступления.

### Демонстрация MVP

**Демонстрация MVP** в рамках проекта - это не готовый сервис, а воспроизводимый прогон вашей системы на заранее подготовленном наборе примеров.

Практически это делается так: в ноутбуке (или через небольшой скрипт) вы заводите блок `Demo`, где лежит 10-15 фиксированных кейсов (`demo_cases.jsonl`), и по ним последовательно показываются входные данные, ключевые промежуточные шаги пайплайна (например, какие документы/чанки были найдены в RAG, какой формат/валидатор сработал, какие правила применились), а затем финальный ответ системы; дополнительно включите 1–2 краевых случая и/или 1 anti-hack кейс (prompt-injection/форматный хак/конфликт источников).

Этого достаточно, чтобы на защите быстро и прозрачно показать, что MVP реально работает, где он ломается и как вы это диагностируете - без серверной части и полноценного UI.

Если вы беспокоитесь за вопроизводимость на защите, можно добавить запись/гифку в презентацию.

## Трек 2 - Воспроизведение статьи

Вы выбираете статью по LLM/NLP и *самостоятельно реализуете* описанный метод и/или воспроизводите ключевые эксперименты (в масштабе, который реально поднять). Цель - показать понимание, инженерную аккуратность и критический анализ.

Особенно рекомендуется выбрать этот трек, если вы пишите **исследовательскую ВКР** по NLP/LLM.


### Выбор статьи (обязательно согласовать)

Статья должна быть по тематике курса: архитектуры LLM, дообучение, preference-alignment (DPO/ORPO/RLHF), интерпретируемость, RAG, оценка, prompting, агенты, безопасность.

Ориентируйтесь на свежие работы: 2024-2026 годы. Если работа еще в preprint и/или на рецензии, вы всё равно можете её использовать.

### Реализация

Нельзя просто запустить авторский репозиторий. Код должен быть ваш (можно опираться на идеи/псевдокод). Если у авторов есть код - сравните с вашей реализацией, объясните расхождения.

### Эксперименты и анализ

Воспроизведите ключевой график/таблицу/вывод (или их упрощенный вариант). Сравните с оригиналом: совпало/не совпало, почему. Добавьте маленькое своё: абляция, другой датасет, другой бейзлайн, или анализ ошибок.

### Отчет

Отчет здесь должен выглядеть как короткий research note **на 1-2 страницы**: чтобы по нему можно было понять, что за статья, что именно вы реализовали, и насколько у вас получилось воспроизвести ключевые результаты. Его можно оформить либо отдельным `report.md` в репозитории, либо как отдельный раздел в ноутбуке (markdown-ячейки).

Вот рекомендованная структура (прям как шаблон):

1. **Реквизиты статьи**

   Название, авторы, год, ссылка. 2-3 предложения: какая проблема и основной вклад.

2. **Что вы воспроизвели**

   Какая часть метода реализована, что упрощали по сравнению с оригиналом. Почему это разумно в рамках ресурсов.

3. **Экспериментальная установка**

   Данные (какие, сколько, как делили), модель/пайплайн, ключевые гиперпараметры, метрики. Пара строк про окружение.

4. **Результаты**

   Одна компактная таблица или 1-2 графика: ваши числа vs числа из статьи, *если возможно*. Если не совпало - это нормально, но нужно явно написать, в каком направлении и на сколько.

5. **Анализ расхождений и ограничения**

   Почему могло не совпасть: отличия данных, токенизации, препроцессинга, сидов, реализации, вычислительных ограничений. Что у метода слабое место на практике.

6. **Что вы добавили от себя (минимум 1 пункт)**

   Одна маленькая добавка: абляция, другая модель, другой датасет, стресс-тест, анализ ошибок, проверка устойчивости.

7. **Вывод (5-8 строк)**

   Коротко: что подтвердили/не подтвердили, чему научились, что бы сделали дальше.

### Формат сдачи и защита

Сдача здесь максимально простая: один аккуратно оформленный ноутбук с экспериментами и короткий отчёт, плюс короткая структурированная защита, где вы проходите по логике статьи и вашим воспроизведённым результатам.

* Ноутбук `.ipynb` + отчет (внутри или отдельно).
* Защита 10 минут: идея статьи -> что реализовали -> результаты -> выводы/ограничения.

## Что допускается и приветствуется

* Использовать LLM-ассистентов (ChatGPT/Copilot и т. п.) можно и приветствуется, если вы понимаете каждый шаг и можете объяснить его на защите.
* Использовать библиотеки и фреймворки (Transformers, LangChain/LlamaIndex, VLLM, RAGAS и т. п.) - можно.
* Брать за основу туториалы/шаблоны (рекомендуем заглянуть в [kaggle code](https://www.kaggle.com/code)) - можно, если вы адаптировали под свою задачу, описали изменения и сделали собственную оценку.
* Для RAG/без-GPU сценариев - можно использовать API-модели; просто фиксируйте модель/параметры и сохраняйте промпты.

## FAQ

**Можно ли работать в команде?**

Нет, данная работа индивидуальная.

**Точно-точно нельзя в команде?**

Если проект реально обуславливает парное участие, то можно. :)

Два условия: а) это согласовывается с преподавателем курса, б) от команды ожидается больше глубины (например, более аккуратная верификация или дополнительная абляция).

**Что если у меня нет GPU?**

Для RAG, промпт-пайплайнов, мини-бенчмарков GPU не нужен. Для дообучения - вы можете использовать бесплатные ресурсы Colab/Kaggle.

Главное - воспроизводимость: понятные зависимости, фиксированные параметры, адекватный README.

**Насколько рабочим должен быть MVP?**

Он должен запускаться и показывать результат на реальных примерах. Идеальное качество не требуется - важны корректный дизайн, проверка качества и честный анализ отказов.

**Что значит "самостоятельная реализация" (Трек 2)?**

Ваш ноутбук должен быть написан вами: можно подсматривать в авторский код, но нельзя просто скопировать и перезапустить.
